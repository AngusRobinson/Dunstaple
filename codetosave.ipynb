```python
#Generating set of diatonic notes for each key signature
tmpkeys = {**reference.all_keysigs["sharps"],**reference.all_keysigs["flats"]}
diatonics = dict.fromkeys(list(tmpkeys.keys()))
diatonics["0s"] = (40,42,44,45,47,49,51)

for key in diatonics:
    if key[1] == "s":
        diatonics[key] = [x + int(key[0])*7 for x in diatonics["0s"]]
    if key[1] == "f":
        diatonics[key] = [x - int(key[0])*7 for x in diatonics["0s"]]

for key in diatonics:
    diatonics[key] = set(filter(lambda x: x>0 and x<88,list(x - (i // 7) * 12 for i, x in enumerate(diatonics[key]*8))+list(x + (i // 7) * 12 for i, x in enumerate(diatonics[key]*8))))
```
